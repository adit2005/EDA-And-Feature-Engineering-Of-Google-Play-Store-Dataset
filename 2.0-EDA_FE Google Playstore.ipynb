```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

```python
game_category = df_copy[df_copy['Category'] == 'GAME']

# Group by app names and sum the installs
game_apps = game_category.groupby('App')['Installs'].sum().sort_values(ascending=False).reset_index()
game_apps.Installs = game_apps.Installs / 1000000000  # converting into billions

# Select the top 10 game apps
top_game_apps = game_apps.head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x='Installs', y='App', data=top_game_apps)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('Game Apps')
ax.set_title("Most Popular Game Apps in Play Store", size=20)

plt.show()
```

```python
communication_category = df_copy[df_copy['Category'] == 'COMMUNICATION']

# Group by app names and sum the installs
communication_apps = communication_category.groupby('App')['Installs'].sum().sort_values(ascending=False).reset_index()
communication_apps.Installs = communication_apps.Installs / 1000000000  # converting into billions

# Select the top 10 communication apps
top_communication_apps = communication_apps.head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x='Installs', y='App', data=top_communication_apps)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('Communication Apps')
ax.set_title("Most Popular Communication Apps in Play Store", size=20)

plt.show()
```

```python
tools_category = df_copy[df_copy['Category'] == 'TOOLS']

tools_apps = tools_category.groupby('App')['Installs'].sum().sort_values(ascending=False).reset_index()
tools_apps.Installs = tools_apps.Installs / 1000000000

top_tools_apps = tools_apps.head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x='Installs', y='App', data=top_tools_apps)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('Tools Apps')
ax.set_title("Most Popular Tools Apps in Play Store", size=20)

plt.show()
```

```python
productivity_category = df_copy[df_copy['Category'] == 'PRODUCTIVITY']

productivity_apps = productivity_category.groupby('App')['Installs'].sum().sort_values(ascending=False).reset_index()
productivity_apps.Installs = productivity_apps.Installs / 1000000000

top_productivity_apps = productivity_apps.head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x='Installs', y='App', data=top_productivity_apps)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('Productivity Apps')
ax.set_title("Most Popular Productivity Apps in Play Store", size=20)

plt.show()
```

```python
social_category = df_copy[df_copy['Category'] == 'SOCIAL']

social_apps = social_category.groupby('App')['Installs'].sum().sort_values(ascending=False).reset_index()
social_apps.Installs = social_apps.Installs / 1000000000

top_social_apps = social_apps.head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x='Installs', y='App', data=top_social_apps)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('Social Apps')
ax.set_title("Most Popular Social Apps in Play Store", size=20)

plt.show()
```
